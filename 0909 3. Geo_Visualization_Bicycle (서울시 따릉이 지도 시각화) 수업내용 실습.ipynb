## 지도시각화(클러스터 & 미니맵)

```python
from folium.plugins import MiniMap
from folium.plugins import MarkerCluster 

# from folium.plugins import MiniMap, MarkerCluster  (위와 동일)
```

```python
### 따릉이 지도 만들기 2 ### 

# 지도 생성하기

m = folium.Map(location = [37.5536067, 126.9674309], zoom_start = 13)

# 미니맵 추가하기
minimap = MiniMap()
minimap.add_to(m)


# 마커 클러스터 만들기
  # 클러스터 추가하기
marker_cluster = MarkerCluster()
m_cluster = marker_cluster.add_to(m) # 클러스터는 꼭! 이름 할당해주고 나중에 지도에 추가해야한다 

    
    
# 마커 추가하기

for i in df.index :
    lat = df.iloc[i].stationLatitude
    long = df.iloc[i].stationLongitude
    name = df.iloc[i].stationName
    available = df.iloc[i].parkingBikeTotCnt
    total = df.iloc[i].	rackTotCnt	 

    if available/total >0.5:
        color = 'blue'
    elif available <2 :
        color = 'red'
    else :
        color = 'green'
        
    icon = folium.Icon(color = color, icon = 'info-sign')
    folium.Marker(location = [lat, long],
                  tooltip = f'{name} : {available}',
                  icon = icon).add_to(m_cluster)

m
    
    
    
    # 자전거 수량에 대해 색상으로 표시
    ##  자전거 보유율이 50% 초과일 경우 --> 파란색
    ##  현재 자전거가 2대 보다 적을 경우 --> 빨간색
    ##  그 외의 경우(자전거 2대 이상 이면서, 자전거 보유율 50% 미만) --> 초록색
   

        
#     print(name, available, total, lat, long)
```
